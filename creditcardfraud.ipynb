```python
#FROM PLOTTING WE CAN SEE THE DISTRIBUTION OF AMOUNT FOR FRAUD AND NORMAL TRANSACTIONS

def plot_hist(data1, bins,column,log_scale = True):
    bins = 100
    fraud = data1[data1.Class == 1]
    normal = data1[data1.Class == 0]
    fig,(ax1,ax2) = plt.subplots(2,1,sharex = True, figsize=(6,4))
    fig.suptitle(f"counts of {column} transaction")
    
    ax1.hist(fraud[column], bins = bins, color = 'r', alpha =0.5, density = True)
    ax1.set_title('fraud')
    
    ax2.hist(normal[column],bins=bins, color = 'g', alpha = 0.5, density = True)
    ax2.set_title('normal')

    
    plt.xlabel(column)
    plt.ylabel("counts")

    
    if log_scale:
        plt.yscale('log')

    plt.xlim((np.min(data1[column]), np.max(data1[column])))
    plt.show()
```

```python
bins = 100
plot_hist(data1, bins, 'Amount')
```

```python
#THIS SHOWS THAT THE FRAUD TRANSACTIONS DONT HAVE HIGH VALUES AT ALL
#LET UNDERSTAND BY ANOTHER VISUALIZATION BY BOXPLOTTING
fig, (ax1,ax2) = plt.subplots(2,1, figsize = (6,4), sharex = True)
fig.suptitle('Boxplot of transaction Amount')

ax1.boxplot(fraud.Amount, vert = False, labels = ['fraud'])
ax1.set_title('fraud')

ax2.boxplot(normal.Amount, vert = False, labels = ['normal'])
ax2.set_title('normal')
```

```python
#THERE ARE OUTLIERS WITH HEAVEY AMOUNT IN NORMAL DATASET
fraud_outliers = fraud[fraud.Amount > 10000]
normal_outliers = normal[normal.Amount > 2500]

print(f"Fraud outliers : {fraud_outliers.shape}")
print(f"Normal outliers : {normal_outliers.shape}")
```

```python
#COMPARISON BETWEEN FRAUD AMOUNT AND NORMAL AMOUNT
fig,(ax1,ax2) = plt.subplots(2,1, figsize = (6,4), sharex = True)
fig.suptitle('Transaction Amount Comparison')

ax1.scatter(fraud.index, fraud.Amount, color = 'r', alpha = 0.5)
ax1.set_title('Fraud')

ax2.scatter(normal.index, normal.Amount, color = 'y', alpha = 0.5)
ax2.set_title('Normal')

plt.xlabel('Transaction index')
```

```python
fig,ax = plt.subplots(1,2, figsize = (15,6))
fig.suptitle('Comparison between fraud and normal transaction wrt amount')

sns.kdeplot(normal["Amount"], label = "Amount", ax = ax[0])

sns.kdeplot(fraud["Amount"], label = "Amount", ax = ax[1])


plt.show()
```

```python
#FOR LOW AMOUNT OF TRANSACTION , THE PROBABILITY OF FRAUD IS MORE THAN A REGULAR TRANSACTION, BUT MOST OF THE TRANSACTION ARE NORMAL

fig,ax = plt.subplots(9,3, figsize = (40,40))
ax = ax.flatten()

for i,column in enumerate(data1.columns[1:28]):
    sns.boxplot(x = 'Class', y = column, data = data1, ax = ax[i])

    plt.xlabel('Class')
    plt.ylabel(column)
    plt.tight_layout()

    fig.suptitle('Fraud Vs Normal Transactions wrt Features')
```

```python
#FEATURE BY CORRELATION

plt.figure(figsize=(30,20))
sns.heatmap(data1.corr(), cbar = True, annot = True, cmap = 'coolwarm', fmt = '.2f')
```

```python
#DROPPING THE COLUMNS HAVING CORRELATION LESS THAN 0.13

for corr,column in zip(data1.corr().iloc[30],data1.columns):
    if abs(corr) < 0.13:
        print(f"{column} has a correlation with class: {corr}")

        data1.drop(columns=column,inplace = True)
```

```python
#HEATMAP OF MODIFIED DATAFRAME

plt.figure(figsize=(15,6))
plt.suptitle('Heatmap Of Modified Dataframe')
sns.heatmap(data1.corr(), cbar = True, annot = True, cmap = 'coolwarm', fmt = '.2f')
plt.show()
```
